```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv("customer_segmentation.csv")

print(data.head())
```

```python

data_num = data.select_dtypes(include=['int64','float64'])

data_num = data_num.drop("ID", axis=1)

data_num = data_num.fillna(data_num.mean())
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(data_num)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)

kmeans_labels = kmeans.fit_predict(X)

print("K-Means Labels:")
print(kmeans_labels[:10])

k_score = silhouette_score(X, kmeans_labels)
print("K-Means Silhouette Score:", k_score)
```

```python
hier = AgglomerativeClustering(n_clusters=4)

hier_labels = hier.fit_predict(X)

print("Hierarchical Labels:")
print(hier_labels[:10])

# checking performance
h_score = silhouette_score(X, hier_labels)
print("Hierarchical Silhouette Score:", h_score)
```

```python
data_num["KMeans_Cluster"] = kmeans_labels
data_num["Hier_Cluster"] = hier_labels
```

```python
print("K-Means Cluster Summary")
print(data_num.groupby("KMeans_Cluster").mean())

print("\nHierarchical Cluster Summary")
print(data_num.groupby("Hier_Cluster").mean())
```
